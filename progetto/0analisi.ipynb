```python
def listToString(s): 
    # initialize an empty string
    str1 = "" 
    # traverse in the string  
    for ele in s: 
        str1 += str(ele)  
    # return string  
    return str1 

def stringToList(string):
    ris=[]
    for elem in string:ris.append(int(elem))
    return ris
    
def getEsamiSuperati (studente:pd.DataFrame):
    esamiSuperati = np.zeros(7,dtype=int)
    for i in range(len(studente)):
        voto = studente.iloc[i,0]
        if voto != 0 : 
            esamiSuperati[i]=1
    return esamiSuperati

def getCombinazioni(arrayDiArray):
    combinazioni = []
    conteggi = []
    for elem in arrayDiArray:
        elem = listToString(elem)
        if not elem in combinazioni:
            combinazioni.append(elem)
            conteggi.append(1)
        else:
            indice = combinazioni.index(elem, 0, len(combinazioni))
            conteggi[indice]+=1
    ris = []
    for elem in combinazioni:
        ris.append(stringToList(elem))
    return ris,conteggi
```

```python
labels = ['BNB + GBC*', 'random forest class.', 'decision tree', 'SVC', 'naive bayes', 'nearest neighbour', 'SGD', 'regressione lineare', 'XGBR + XGBR + RFR*', 'regressione logistica', 'elastic net regr.*','neural network']
mesi = ['gen','feb','mar','apr','mag','giu','lug','ag','set','ott','nov','dic']
#acc[0]=gennaio, acc[2]=febbraio, ...
accuracy = []
precision = []
recall = []
f1= []

larghezzaGrafici = 1.2
altezzaGrafici = 1
linewidth = 3
```

```python
#lettura dal file 'performance1.txt
coseLette = []
with open("performance1.txt") as f:
    for line in f: 
        line = line.strip() #or some other preprocessing
        coseLette.append(line) #storing everything in memory!

#per ogni riga del file
for i in range(len(coseLette)):

    temp = []
    # spezzo la stringa in float
    coseLetteSplit = coseLette[i].split()
    for j in range(len(coseLetteSplit)):coseLetteSplit[j]=float(coseLetteSplit[j])

    for elem in coseLetteSplit:
            temp.append(elem)

    if i % 4 == 0:
        accuracy.append(temp.copy())
    if i % 4 == 1:
         precision.append(temp.copy()) 
    if i % 4 == 2:
         recall.append(temp.copy())
    if i %4 ==3:
         f1.append(temp.copy())
```

```python
#ottengo array di array, ciascuno con le performance di un modello diverso
accuracyModelli = []
precisionModelli = []
recallModelli = []
f1Modelli= []

#per colonna
for i in range(len(accuracy[0])):
    temp = [[],[],[],[]]
    #per ogni elemento della colonna
    for j in range(len(accuracy)):
        temp[0].append(accuracy[j][i])
        temp[1].append(precision[j][i])
        temp[2].append(recall[j][i])
        temp[3].append(f1[j][i])

    accuracyModelli.append(temp[0])
    precisionModelli.append(temp[1])
    recallModelli.append(temp[2])
    f1Modelli.append(temp[3])
```

```python
#MODELLO 1 
#-----------------------------------------accuracy
print('accuracy')

myFontSize = 15
height = 0.2

fig=plt.figure()
ax = fig.add_axes([0.1, 0.1,larghezzaGrafici,altezzaGrafici]) # main axes
ax.set_xlabel('mesi',fontsize=myFontSize)
ax.set_ylabel('accuracy',fontsize=myFontSize)
ax.grid(color='k', linestyle='-', linewidth=1, axis='x', alpha=0.2)
ax.grid(color='k', linestyle='-', linewidth=1, axis='y', alpha=0.2)

daNonStampare = [
    'naive bayes','SGD','SGD','regressione logistica','regressione lineare','nearest neighbour','BNB + GBC*', 'SVC'
]
colors = [BLU,GIALLO,VERDE,VIOLA,'c']
indiceColore = 0
for i in range(len(accuracyModelli)):
    if labels[i] not in (daNonStampare):
        ax.plot(mesi,accuracyModelli[i],label = labels[i],linewidth=linewidth,color = colors[indiceColore])
        indiceColore+=1
#stampo gradient boosting classifier per ultimo
ax.plot(mesi,accuracyModelli[0],label = labels[0],linewidth=linewidth,color='r')

ax.tick_params(axis='both',labelsize = myFontSize)
ax.set_xticklabels(mesi)

ax.legend(prop={'size': 14})
ax.plot
plt.savefig('generatedImages/accuracyModello1.png', dpi=300,bbox_inches='tight',facecolor='w')
```
